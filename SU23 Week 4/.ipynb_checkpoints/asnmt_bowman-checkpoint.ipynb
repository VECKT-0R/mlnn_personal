
## 2a. Create 3 models fit to (`x=hour`, `y=monday`)  with varying polynomial degrees ( choose from `n=5,15,20`). (Repeat for `saturday` below)

## Plot all the results for each polynomial.

```python
from sklearn import linear_model, metrics
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

linear = linear_model.LinearRegression()

ridge = linear_model.Ridge()

n_5 = PolynomialFeatures(degree = 5)
n_15 = PolynomialFeatures(degree = 15)
n_20 = PolynomialFeatures(degree = 20)
```

```python
# Monday

monday_clean = monday.dropna()

mon_x = monday_clean.hour

mon_y = monday_clean.monday
```

```python
monday_5 = n_5.fit_transform(mon_x.values.reshape(-1,1))

linear.fit(monday_5, mon_y)

#ridge.fit(monday_5, mon_y) I tested ridge and it is extremely similar to linear in this case

plt.figure(figsize=(20,10))
plt.plot(mon_x, mon_y, c = "r")
plt.plot(mon_x, np.dot(monday_5, linear.coef_) + linear.intercept_, c = "purple")
#plt.plot(mon_x, np.dot(monday_5, ridge.coef_) + ridge.intercept_, c = "cyan")
```

```python
monday_15 = n_15.fit_transform(mon_x.values.reshape(-1,1)) #Probably best model of these cases

linear.fit(monday_15, mon_y)

#ridge.fit(monday_15, mon_y) I tested ridge and its significantly worse than linear

plt.figure(figsize=(20,10))
plt.plot(mon_x, mon_y, c = "r")
plt.plot(mon_x, np.dot(monday_15, linear.coef_) + linear.intercept_, c = "purple")
#plt.plot(mon_x, np.dot(monday_15, ridge.coef_) + ridge.intercept_, c = "cyan")
```

```python
monday_20 = n_20.fit_transform(mon_x.values.reshape(-1,1))

linear.fit(monday_20, mon_y)

#ridge.fit(monday_20, mon_y) I tested ridge and its significantly worse than linear

plt.figure(figsize=(20,10))
plt.plot(mon_x, mon_y, c = "r")
plt.plot(mon_x, np.dot(monday_20, linear.coef_) + linear.intercept_, c = "purple")
#plt.plot(mon_x, np.dot(monday_20, ridge.coef_) + ridge.intercept_, c = "cyan")
```

## 2b. Repeat `2a` for `saturday`
